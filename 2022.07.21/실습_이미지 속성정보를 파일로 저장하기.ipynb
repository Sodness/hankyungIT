```python
# 이미지 자르기 (crop)

from PIL import Image
import os
import pandas as pd
from glob import glob

path_get = './images/'
path_put = './images/crop/'
files = os.listdir(path_get)
# print(files)

# 파일명 가져오기 : os.listdir(폴더경로)
# 폴더 생성:os.mkdir(폴더명)
# 폴더가 없으면 생성
# 디렉토리 있는지 확인: os.path.isdir
if os.path.isdir(path_put[0:-1]) == False:
    print('mkdir')
    os.mkdir(path_get + 'crop')

files = glob(path_get + '*.jpg')

# 이미지 자르기
for file in files:
    print(file)
    img = Image.open(file)
    name = img.filename.split('\\')[-1]
    name = name.split('.')[0]
    # print(name)
    
    # if 사이즈변경후 사진의 이름 in resize폴더 안에 있으면:
    if f'{name}_400x400.jpg' in os.listdir(path_put):
        print('break')
        break
    
    cropped_img = img.crop((100, 100, 200, 200))
    cropped_img.save(path_put + f'{name}_400x400.jpg')
```

```python
# 이미지 대칭 (flip)

from PIL import Image
import os
import pandas as pd
from glob import glob

path_get = './images/'
path_put = './images/flip/'

# 파일명 가져오기 : os.listdir(폴더경로)
# 폴더 생성:os.mkdir(폴더명)
# 폴더가 없으면 생성
# 디렉토리 있는지 확인: os.path.isdir
if os.path.isdir(path_put[0:-1]) == False:
    print('mkdir')
    os.mkdir(path_get + 'flip')

files = glob(path_get + '*.jpg')
# print(files)

# 이미지 자르기
for file in files:
    print(file)
    img = Image.open(file)
    name = img.filename.split('\\')[-1]
    name = name.split('.')[0]
    # print(name)
    
    # if 사이즈변경후 사진의 이름 in resize폴더 안에 있으면:
    if f'{name}_flip.jpg' in os.listdir(path_put):
        print('break')
        break
    
    flipped_img = img.transpose(Image.FLIP_LEFT_RIGHT)
    flipped_img.save(path_put + f'{name}_flip.jpg')
```

```python
# 이미지 필터링 (filter)

from PIL import Image
from PIL import ImageFilter
import os
import pandas as pd
from glob import glob

path_get = './images/'
path_put = './images/filter/'

# 파일명 가져오기 : os.listdir(폴더경로)
# 폴더 생성:os.mkdir(폴더명)
# 폴더가 없으면 생성
# 디렉토리 있는지 확인: os.path.isdir
if os.path.isdir(path_put[0:-1]) == False:
    print('mkdir')
    os.mkdir(path_get + 'filter')

files = glob(path_get + '*.jpg')
# print(files)

# 이미지 자르기
for file in files:
    print(file)
    img = Image.open(file)
    name = img.filename.split('\\')[-1]
    name = name.split('.')[0]
    # print(name)
    
    # if 사이즈변경후 사진의 이름 in resize폴더 안에 있으면:
    if f'{name}_filter.jpg' in os.listdir(path_put):
        print('break')
        break
    
    filtered_img = img.filter(ImageFilter.GaussianBlur(5))
    filtered_img.save(path_put + f'{name}_filter.jpg')
```
