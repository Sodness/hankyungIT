```python
data = [
    [ 2000,  3050,  2050,  1980],
    [ 7500,  2050,  2050,  1980],
    [15450, 15050, 15550, 14900]
]

res = []
for i in data:
    for j in i:
        res.append(j * 1.00014)
print(res)
```

```python
data = [
    [ 2000,  3050,  2050,  1980],
    [ 7500,  2050,  2050,  1980],
    [15450, 15050, 15550, 14900]
]

res = []
for i in data:
    sub = []
    for j in i:
        sub.append(j * 1.00014)
    res.append(sub)
print(res)
```
