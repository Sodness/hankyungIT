## 강사님 주피터 노트북 꺼져서 복습차원에서 다시

```python
df = pd.read_csv('./data/cctv_in_seoul.csv')
df2 = df.copy()
```

```python
df2.head()
```

```python
df2['2016년'] - df2['2015년']
```

```python
df2['구분'] = ['상' if i>=500 else '중' if i>=300 else '하' for i in df2['2016년']]
df2
```

```python
df2['2016년'].quantile([0.25, 0.5, 0.75])
```

```python
mn,c,mx = df2['2016년'].quantile([0.25, 0.5, 0.75])
print(mn, c, mx)
```

```python
# IRQ: 박스플롯의 박스부분 최대값 - 최소값
IRQ = mx - mn
max_outlier = mx + (IRQ*1.5)
min_outlier = mn - (IRQ*1.5)
df2[df2['2016년'] >= max_outlier]
```

```python
def max_o(x):
    mn,c,mx = x.quantile([0.25, 0.5, 0.75])
    IRQ = mx - mn
    max_outlier = mx + (IRQ*1.5)
    return df2[x >= max_outlier]
```

```python
max_o(df2['2014년'])
```

```python
max_o(df2['2015년'])
```

```python
def max_o_s(s_df, x):
    mn,c,mx = s_df[x].quantile([0.25, 0.5, 0.75])
    IRQ = mx - mn
    max_outlier = mx + (IRQ*1.5)
    return s_df[s_df[x] >= max_outlier]
```

```python
max_o_s(df2, '2016년')
```

```python
def min_o_s(s_df, x):
    mn,c,mx = s_df[x].quantile([0.25, 0.5, 0.75])
    IRQ = mx - mn
    min_outlier = mn - (IRQ*1.5)
    return s_df[s_df[x] <= min_outlier]
```

```python
min_o_s(df2, '2016년')
```
