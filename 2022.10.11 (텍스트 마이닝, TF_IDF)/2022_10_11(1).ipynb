```python
import pandas as pd
from konlpy.tag import Okt
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/전처리_샘플뉴스.csv')
df
```

```python
df['text'][0]
```

```python
okt=Okt()
words=okt.nouns(df['text'][0])
str_words_1=' '.join(words)
str_words_1
```

```python
okt=Okt()

res = []
for i in df['text']:
    words=okt.nouns(i)
    str_words=' '.join(words)
    res.append(str_words)
```

```python
res[0]
```

```python
str_words=' '.join(res)
```

```python
stopwords = ['명', '달', '이', '등', '고', '중', '것', '수', '씨', '며', '건', '말']
```

```python
len(str_words_1)
```

```python
txt = str_words_1

wc = WordCloud('./NanumGothic.ttf',
               mode='RGBA',
               background_color = None,
               max_words = 150,
               stopwords = stopwords)
wc = wc.generate(txt) # str자료만 가능, list자료는 안됨
# plt.figure(figsize = (12, 12))
plt.imshow(wc) #interpolation = "bilinear")
plt.axis("off")

# plt.savefig('정책브리핑_태양광_워드클라우드.png')
plt.show()
```
