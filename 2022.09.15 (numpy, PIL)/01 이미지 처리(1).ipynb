### 1. numpy 형변환
- np변수.reshape((shape))
- np변수.flatten()

```python
# np.reshape((shape))
import numpy as np
tmp=np.arange(24)
print(tmp)
print('---------------------------')
print(tmp.reshape((4,6)))
print(tmp.reshape((6,4)))
print(tmp.reshape((6,4,1)))
print(tmp.reshape((1,6,4)))
print('-------------------------------')
print(tmp.reshape((1,6,-1))) #-1 옵션
```

```python
### np변수.flatten => 1차원으로 변경
img=np.zeros((8,8,3))
print(img[:2])
print(img.reshape((24,-1))[:2])
#respahe로 1차원으로 변경
print(img.reshape(-1))

#flattern으로 1차원으로 변경
print(img.flatten())
```

### 이미지 자료 브로드캐스트
- 이미지를 array화 시키면 [행,열,[rgb]] 
- PILLOW 모듈은 np.array로 해야하고, openCV모듈은 읽을때 array화 되어 있음.
    - [:,:,0] =>RGB의 R값임
    - [:,:,1] =>RGB의 G값임
    - [:,:,2] =>RGB의 B값임
    - [...]은 행,열,rgb 뜻임 =[:,:,:] 

```python
import numpy as np
import matplotlib.pyplot as plt

img=[[[255,0,0],[0,255,0]],
     [[0,0,255],[255,255,0]],
     [[255,0,255],[0,255,0]],
     [[255,0,0],[0,255,255]]]
np.shape(img)
plt.imshow(img)
```

```python
# 참고: 이미지 자료는 화면에서 play하는 용도는 0~255나, 0~1이나 같음.
# 이미지 정규화(Normalization)임. 이 이미지 정규화(Normalization)은 작업한 사람밖에 모름
# 인공지능할때 255로 나누었는지 127로 나누었는지 모름.
# 인공지능 테스트할때는 정규화(Normalization)하고 똑같이 해야함.  255로 나누면 새 이미지로 255로 나눠야함.
# 그래서 작성한 사람이 안 알려주면 모름. 결과 잘 안나옴
# 인공지능 작업할때는 상황에 따라 255로, 127, 나누던지 127로 나누고 -1 하던지 그때 그때 이미지 상황에 따라서 
# 작업이 달라지고, 정답은 없음.
# -그러나 0~255 보다는 -1~1 또는 0~1이 값이 작음으로 이미지 정규화(Normalization)을 권장함.
# 그래서 numpy로 받은 자료(npy나 npz 임)는 반드시 min, max를 확인해야함.

img=[[[255,0,0],[0,255,0]],
     [[0,0,255],[255,255,0]],
     [[255,0,255],[0,255,0]],
     [[255,0,0],[0,255,255]]]


#img
plt.subplot(221)
img=np.array(img)
plt.imshow(img)
plt.title('img')

#img/255
plt.subplot(222)
img_1=img/255
plt.imshow(img_1)
plt.title('img/255')

#img/127
plt.subplot(223)
img_2=img/127
plt.title('img/127')
plt.imshow(img_2)

#img/127-1
plt.subplot(224)
img_3=img/127-1
plt.imshow(img_3)
plt.title('img/127-1')
plt.tight_layout()
plt.show()

print('img:',np.min(img), np.max(img), np.shape(img))
print('img/255:',np.min(img_1), np.max(img_1), np.shape(img_1))
print('img/127:',np.min(img_2), np.max(img_2), np.shape(img_2))
print('img/127-1:',np.min(img_3), np.max(img_3), np.shape(img_3))
```

### 사진1장 확장법
- resahpe은 인공지능할때 필수 임.
- 인공지능에서는 예측할때 반드시 (갯수,행,열,채널수)  # 한개의 이미지가 8,8,3 이라면 1,8,8,3으로 변경해야함
- reshape와 같은 기능은 expend_dims와 np.newaxis가 있음

```python
#reshap 활용
# 1장의 이미지(8,8,3) => (1,8,8,3)으로 변경
img=np.zeros((8,8,3))
tmp=img.reshape(1,8,8,3)   # array가 아닌 자료는 np.reshape(img, (1,8,8,3))
np.shape(tmp)
```

```python
# np.newaxis 활용
# 1장의 이미지(8,8,3) => (1,8,8,3)으로 변경

img=np.zeros((8,8,3))
tmp=img[np.newaxis,...]
print(tmp.shape)
tmp=img[...,np.newaxis]
print(tmp.shape)
```

```python
# np.expand_dims 활용
# 1장의 이미지(8,8,3) => (1,8,8,3)으로 변경
img=np.zeros((8,8,3))
tmp=np.expand_dims(img,axis=0)
print(tmp.shape)
```

```python
# 4채널 있어서 에러 나옴
from PIL import Image
from glob import glob
fileList=glob('./img/*.jpg')
result=[]
for x in fileList:
    img=Image.open(x).resize((100,100))
    
    img=np.array(img)
    print(img.shape)
    result.append(img)
np.shape(result)
```

```python
## 3채널만 append함
## 이렇게 넘파이시킨자료는 npy 자료로 저장가능함
from PIL import Image
from glob import glob
fileList=glob('.\\img\\*.jpg')
result=[]

for x in fileList:
    img=Image.open(x).resize((100,100))
    
    img=np.array(img)/127
    if img.shape[2]==3:
   
        result.append(img)
result
```

```python
for cnt,x in enumerate(result):
    plt.subplot(1,7,cnt+1)
    plt.imshow(x)
```

```python
plt.imshow(np.hstack((result[0],result[1],result[2])))
plt.show()
plt.imshow(np.hstack((result[:3])))
```
