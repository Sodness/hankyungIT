#### 통계 함수
 - 평균, 분산, 중앙, 최대, 최소값 등등 통계 관련된 함수가 내장

```python
print(y)
```

```python
#평균
np.mean(y)
```

```python
y.mean()
```

```python
#최대값, 최소값
np.max(y), np.min(y)
```

```python
print(np.argmax(y)) #최대값의 위치 인덱스
print(np.argmin(y)) #최대값의 위치 인덱스
```

```python
#분산, 중앙값,표준편차
np.var(y), np.median(y), np.std(y)
```

#### any, all 함수
 - any: 특정 조건을 만족하는 것이 하나라도 있으면 True, 아니면 False
 - all: 모든 원소가 특정 조건을 만족한다면 True, 아니면 False

```python
z = np.random.randn(10)
print(z)
```

```python
z > 0
```

```python
np.any(z > 0)
```

```python
np.all(z != 0)
```

#### where 함수
 - 조건에 따라 선별적으로 값을 선택 가능
 - 사용 예) 음수인경우는 0, 나머지는 그대로 값을 쓰는 경우

```python
z = np.random.randn(10)
print(z)
```

```python
np.where(z > 0, z, 0)
```

```python
np.where(z > 0,z,0)
```
