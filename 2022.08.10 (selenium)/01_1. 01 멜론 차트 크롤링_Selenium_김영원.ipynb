```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import time
```

```python
driver = webdriver.Chrome('./chromedriver.exe')
url = 'https://www.melon.com/chart/day/index.htm'
driver.get(url)
time.sleep(2)
html = driver.page_source
driver.close()    # 현재 탭을 종료
driver.quit()     # 브라우저를 종료
```

```python
html
```

```python
type(html)
```

```python
soup = bs(html, 'html.parser')
```

```python
soup
```

```python
type(soup)
```

```python
tr_tags = soup.select('form#frm div table tbody tr')
```

```python
len(tr_tags)
```

```python
for tr_tag in tr_tags:
    rank = tr_tag.find('span', 'rank').text
    print(tr_tag)
    break
```

```python
date = soup.find('span', 'year').text

res = []
for tr_tag in tr_tags:
    rank = tr_tag.find('span', 'rank').text
    title = tr_tag.find('div', 'ellipsis rank01').find('a').text
    artist = tr_tag.find('div', 'ellipsis rank02').find('a').text
    album = tr_tag.find('div', 'ellipsis rank03').find('a').text
    like = tr_tag.find('span', 'cnt').text.replace('총건수','').strip()
    # print(rank, title, artist, album, like)
    
    dict = {
        '날짜': date,
        '순위': rank,
        '제목': title,
        '가수': artist,
        '앨범': album,
        '좋아요': like
    }
    res.append(dict)
```

```python
df = pd.DataFrame(res)
```
