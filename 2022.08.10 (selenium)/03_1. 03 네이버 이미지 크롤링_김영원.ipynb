```python
# 검색어 = '강아지'
검색어 = input('다운받고 싶은 이미지 입력:')
cnt = int(input('스크롤 횟수:'))
cnt = 6
if cnt >= 6:
    cnt = 6
url = 'https://search.daum.net/search?w=img&nil_search=btn&DA=NTB&enc=utf8&q='+검색어
# print(url)

driver = webdriver.Chrome('./chromedriver.exe')
driver.get(url)
time.sleep(2)

for i in range(cnt):
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight);")
    time.sleep(2)

# print(driver.find_elements(By.CLASS_NAME, '_image'))

elems = driver.find_elements(By.CLASS_NAME, 'thumb_img')
imgs = []
for i in elems:
    imgs.append(i.get_attribute('src'))

driver.close()
driver.quit()

os.makedirs(f'./{검색어}/', exist_ok=True)

for idx,img in enumerate(imgs, 1):
    urllib.request.urlretrieve(img, f'./{검색어}/{검색어}{idx}.jpg')
```
