```python
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import requests
import pandas as pd
from tqdm import tqdm
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://www.korea.kr/totalSearch.do?srchFlag=Y&keyword=%EB%B6%80%EB%8F%99%EC%82%B0&oldKeyword=%EB%B6%80%EB%8F%99%EC%82%B0&resultKeyword=%EB%B6%80%EB%8F%99%EC%82%B0&sort=date&scnt=1&punit=10&where=news_korea_new_v2&dateRange=all&range=ALL&op=and&multiFlag=Y&news_type=pol'
driver = webdriver.Chrome('./chromedriver.exe')
driver.get(url)
time.sleep(2)

# 검색어 입력
# 검색어 = '청년실업'
검색어 = input('정책브리핑 검색어 입력:')
# pages = 14
pages = int(input('검색 페이지 수:'))
current_page = 0

# 검색어 검색
elem = driver.find_element(By.XPATH, '//*[@id="topSearchKeywordSr"]')
elem.clear()
elem.send_keys(검색어)
elem.send_keys(Keys.ENTER)
time.sleep(2)

# 1페이지 담기
driver.find_element(By.XPATH, '//*[@id="news_korea_new_v2"]/div[1]/div/ul/li[2]/a').click()
current_page = 1
time.sleep(2)
res = [bs(driver.page_source, 'html.parser')]

# 나머지 페이지 담기
for page in range(2, pages+1):
    if current_page % 10 == 0:
        driver.find_element(By.XPATH, '//*[@id="paging"]/div/a[3]').click()
        current_page = current_page + 1
        time.sleep(2)
        res.append(bs(driver.page_source, 'html.parser'))
    else:
        driver.find_element(By.LINK_TEXT, str(page)).click()
        current_page = page
        time.sleep(2)
        res.append(bs(driver.page_source, 'html.parser'))

driver.close()
driver.quit()
```

```python
data_ls = []
for i in range(len(res)):
    title_tags = res[i].select('div#result_news_korea_new_v2 ul li span.text strong a')
    for title_tag in title_tags:
        title = title_tag.text.strip()
        href = 'https://www.korea.kr/' + title_tag['href']
        # print(title, href)
        data_ls.append([title, href])

df = pd.DataFrame(data_ls, columns=['제목', '링크'])
```

```python
data_ls2 = []
for href in df['링크']:
    # print(href)
    html = requests.get(href)
    soup = bs(html.text, 'html.parser')
    
    title = soup.select_one('div.article-head h1').text
    date = soup.select('div.info span')[0].text
    writer = soup.select('div.info span')[1].text
    content = soup.select_one('div.view-cont').text
    
    data_ls2.append([title, date, writer, content])
    print(date, writer)
    # break
```

```python
df_href = pd.DataFrame(data_ls2, columns=['제목', '작성일', '작성자', '내용'])
```
