```python
# =input('   :')
# count=int(input('   (:6):'))
# if count>6:
#     count=6

=''
url='https://search.naver.com/search.naver?where=image&sm=tab_jum&query=%EA%B3%A0%EC%96%91%EC%9D%B4'

driver=webdriver.Chrome('./chromedriver.exe')
driver.get(url)
time.sleep(2)

#   
for i in range(6):
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight);")
    time.sleep(2)
elems=driver.find_elements(By.CLASS_NAME,'_image')
imgs=[]
for elem in elems:
    imgs.append(elem.get_attribute('src'))

driver.quit()
```

```python
for idx,img in enumerate(imgs):
    urllib.request.urlretrieve(img, f'.//{idx}.jpg')
```
