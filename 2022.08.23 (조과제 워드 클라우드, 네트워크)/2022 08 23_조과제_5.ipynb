```python
import pandas as pd
```

```python
df = pd.read_csv('./용원님/기상발전_최종본.csv', encoding='euc-kr')
```

```python
df
```

```python
df_group = df.groupby(['지역','일시']).mean()
```

```python
df_group
```

```python
df_group = df_group.reset_index()
```

```python
df_group
```

```python
tmp = []
year = '2021'
for idx,i in enumerate(df_group['일시']):
    if year in i:
        tmp.append(df_group.loc[idx])

df_tmp = pd.DataFrame(tmp).reset_index(drop=True)
df_tmp['지역'].unique()
df_tmp.to_csv(f'./{year} 전국 최종본.csv', encoding='euc-kr')
```
