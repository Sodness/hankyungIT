```python
np.random.seed(100)
x = np.random.randn(1000)
y = np.random.randn(1000)

plt.scatter(x,y)
# 아 이거 시드값이 같은 셀에 있어야 되네. 아닌가?
```

```python
w = 0.2
b = 0.1
y_hat = w*x + b
plt.scatter(x,y)
plt.plot(x,y_hat)

MSE = np.mean((y - y_hat)**2)
plt.title(MSE)
```

```python
# 임의로 w,b 값을 갱신
w = [0.2, 0.1, 0.3, 2]
b = [0.1, 0.1, 0, -0.5]

for w,b in zip(w, b):
    y_hat = w*x + b
    plt.scatter(x,y)
    plt.plot(x,y_hat)

    MSE = np.mean((y - y_hat)**2)
    plt.title(MSE)
    plt.show()
```

```python
# 임의로 w,b 값을 갱신
import errno


w = [0.2, 0.1, 0.3, 2]
b = [0.1, 0.1, 0, -0.5]
cnt = 0
plt.figure(figsize=(12,8))

for w,b in zip(w, b):
    cnt = cnt + 1
    plt.subplot(1,5,cnt)

    y_hat = w*x + b
    plt.ylim([-8, 8])
    plt.scatter(x,y)
    plt.plot(x,y_hat)

    MSE = np.mean((y - y_hat)**2)
    plt.title(MSE)
plt.show()
```

```python
# 임의로 w,b 값을 갱신
w = 2
b = 0.1
cnt = 0
plt.figure(figsize=(20,8))
x = np.random.randn(1000)
y = np.random.randn(1000)
lr = 0.001
epochs = 5

for i in range(epochs):
    cnt = cnt + 1
    plt.subplot(1,5,cnt)

    y_hat = w*x + b
    plt.ylim([-8, 8])
    plt.scatter(x,y)
    plt.plot(x,y_hat)

    err = np.mean((y - y_hat)**2)
    plt.title(err)
    
    w = w - w*err * lr
    b = b - b*err * lr
plt.show()
```
