```python
df.columns=['지역명', '규모구분', '연도', '월', '분양가격']
```

```python
df.head()
```

```python
#df['분양가격']=='  '  None 채우기
df.loc[df['분양가격']=='  ','분양가격']=None
```

```python
df.isnull().sum()
```

```python
#분양가격 ',' 없애기

df['분양가격']=df['분양가격'].str.replace(',','')
```

```python
#'분양가격']=='-' 에 None 입력
df.loc[df['분양가격']=='-','분양가격']=None
```

```python
# 분양가격 notnull 표를 따로 저장함
df_not_null=df[df['분양가격'].notnull()]
```

```python
df_not_null['분양가격']=df_not_null['분양가격'].astype('int')
```

```python
df_not_null.dtypes
```

```python
# 1. df 분양가격에서...  isnull만추출
# 최종미션: df분양가격의 isnull을  연도별, 지역별,규모별, 분양가격 평균으로 채우기
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df_not_null.dtypes
```

```python
df_not_null.shape
```

```python
df_null=df[df['분양가격'].isnull()]
```

```python
df_not_null.head()
```

```python
df_null.head()
```

```python
df_res=pd.concat([df_not_null,df_null])
```

```python
df_res=df_res.sort_index()
```

```python
df_res
```

```python
df_res.columns
```

```python
# 연도별,지역별, 규모구분별 분양가 평균가격을 분양가 null에 채우기
df_g=df_res.groupby(['연도','지역명','규모구분'])

res=[]
for i,data in df_g:

    temp=data.fillna(data.mean())
    res.append(temp)
# df_완성=pd.concat(res)
# df_완성=df_완성.sort_index()
```

```python
df.isnull().sum().sum()
```

```python
for i in res:
    if (i.isnull().sum().sum())>0 :
        display(i)
```

```python
df_완성=pd.concat(res)
```

```python
df_완성.isnull().sum()
```

```python
df_m=df_완성[df_완성['분양가격'].isnull()]
    
```
