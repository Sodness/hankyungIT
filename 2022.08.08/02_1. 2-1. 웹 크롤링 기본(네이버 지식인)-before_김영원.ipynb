```python
len(soup.find_all('div', 'kin_wrap api_ani_send'))
```

```python
div_tags = soup.find_all('div', 'kin_wrap api_ani_send')
```

```python
res = []
for div in div_tags:
    title = div.find('div', 'question_group').text
    href = div.find('a', 'api_txt_lines question_text')['href']
    # print(div.select_one('div.question_group>a')['href'])
    writer = div.find('span', 'name').text
    contents = div.find('div', 'answer_group').text
    date = div.find('span', 'desc')
    
    # print(href)
    tmp = {
        '제목': title,
        '링크': href,
        '작성자': writer,
        '글 내용': contents,
        '작성일': date
    }
    res.append(tmp)
```

```python
df = pd.DataFrame(res)
df
```
