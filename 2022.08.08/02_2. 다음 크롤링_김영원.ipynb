```python
import requests
from bs4 import BeautifulSoup as bs
from urllib import parse
from urllib import request
from urllib.request import urlopen
import pandas as pd
```

```python
# 블로그 제목,제목href, 날짜, 작성자 -> DataFrame으로 (10개)
검색어 = '스타벅스'
url = 'https://search.daum.net/search?w=blog&DA=PGD&enc=utf8&q='+검색어+'&p=1'

html = requests.get(url)
soup = bs(html.text, 'html.parser')
containers = soup.select('.cont_inner')

res = []
for container in containers:
    title = container.select_one('.f_link_b').text
    href = container.select_one('.f_link_b')['href']
    date = container.find('span', 'f_nb date').text
    writer = container.find('div', 'info clear').find('a', 'f_nb').text
    
    dict = {
        '제목': title,
        '링크': href,
        '작성일': date,
        '작성자': writer
    }
    res.append(dict)

pd.DataFrame(res)
```

```python
# 블로그 제목,제목href, 날짜, 작성자 -> DataFrame으로 (100개)
df_res = []
for page in range(1,11):
    검색어 = '스타벅스'
    page = page
    url = 'https://search.daum.net/search?w=blog&DA=PGD&enc=utf8&q='+검색어+'&p='+str(page)

    html = requests.get(url)
    soup = bs(html.text, 'html.parser')
    containers = soup.select('.cont_inner')

    res = []
    for container in containers:
        title = container.select_one('.f_link_b').text
        href = container.select_one('.f_link_b')['href']
        date = container.find('span', 'f_nb date').text
        writer = container.find('div', 'info clear').find('a', 'f_nb').text

        dict = {
            '제목': title,
            '링크': href,
            '작성일': date,
            '작성자': writer
        }
        res.append(dict)

    df = pd.DataFrame(res)
    df_res.append(df)
    
pd.concat(df_res, ignore_index=True)
```

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
# 검색어 = '원피스'
검색어 = input('다음 블로그 검색어 입력:')
검색어 = 검색어.replace(' ', '+')
url = 'https://search.daum.net/search?w=blog&DA=PGD&enc=utf8&q='+검색어+'&p=1'

html = requests.get(url)
soup = bs(html.text, 'html.parser')

div_tags = soup.find_all('div', 'cont_inner')
res = []
for div in div_tags:
    title = div.find('a', 'f_link_b').text
    href = div.find('a', 'f_link_b')['href']
    date = div.find('span', 'f_nb date').text
    contents = div.find('p', 'f_eb desc').text
    writer = div.find('a', 'f_nb').text
    
    tmp = {
        '제목': title,
        '작성일': date,
        '내용': contents,
        '작성자': writer,
        'URL': href
    }
    res.append(tmp)
    
df = pd.DataFrame(res)
display(df)
```

```python
# 1~10페이지 다음 블로그 크롤링 하기
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd

검색어 = input('다음 블로그의 검색어 입력:')
검색어 = 검색어.replace(' ', '+')
pages = int(input('몇 페이지까지:'))

res = []
for page in range(1, pages+1):
    url = 'https://search.daum.net/search?w=blog&DA=PGD&enc=utf8&q='+검색어+'&p='+str(page)
    html = requests.get(url)
    soup = bs(html.text, 'html.parser')

    div_tags = soup.find_all('div', 'cont_inner')
    for div in div_tags:
        try:
            title = div.find('a', 'f_link_b').text
            href = div.find('a', 'f_link_b')['href']
            date = div.find('span', 'f_nb date').text
            contents = div.find('p', 'f_eb desc').text
            writer = div.find('a', 'f_nb').text

            tmp = {
                '제목': title,
                '작성일': date,
                '내용': contents,
                '작성자': writer,
                'URL': href
            }
            res.append(tmp)
        except:
            pass
df = pd.DataFrame(res)
display(df)
```
