```python
from PIL import Image
import os

path = "./0"
file_list = os.listdir(path)

for file in file_list:
    filename = file.split('.')[0]
    im = Image.open('./0/'+file).convert('RGB')
    im.save('./jpg0/'+ filename +'.jpg', 'jpeg')
    print(filename)
```
