```python
import numpy as np
import matplotlib.pyplot as plt 
from PIL import Image
from glob import glob

plt.figure(figsize=(100,100))
fileList=glob('.\\img\\*.jpg')
result=[]
for x in fileList:
    img=Image.open(x).resize((100,100))
    img=np.array(img)/127 - 1
    if img.shape[2]==3:
        result.append(img)
        
plt.figure(figsize=(15,10))
plt.imshow(np.hstack((result)))
plt.show()
np.save('imgs.npy', result)
```

```python
f_ls = glob('./img/*.jpg')
res = []
for i in f_ls:
    img = Image.open(i).resize((100,100))
    img = np.array(img)/127 - 1
    if img.shape[2] == 3:
        res.append(img)
plt.figure(figsize=(15,10))
plt.imshow(np.hstack(res))
plt.show()
np.save('imgs.npy', res)
```

```python
imgs = np.load('imgs.npy')
print(imgs[0].shape)
print(type(imgs[0]))
plt.imshow(imgs[0])
plt.show()
plt.imshow(np.hstack(imgs))
plt.show()
```
