```python
key = '486a45724e686a6b36306e656e5078'
date = '20210801'
url1 = 'http://openapi.seoul.go.kr:8088/'+key
url2 = '/xml/CardSubwayStatsNew/1/1000/'

for date in dates:
    url = url1 + url2 + date
    # print(url)
    
    xml = requests.get(url)
    if xml.status_code == 200:
        print('접속됨')
    else:
        print('접속실패')
        
    soup = bs(xml.text, 'html.parser')
    rows = soup.find_all('row')
    res = []
    ls = ['USE_DT', 'LINE_NUM', 'SUB_STA_NM', 'RIDE_PASGR_NUM', 'ALIGHT_PASGR_NUM']
    
    for row in rows:
        # print(row)
        dt = {}
        for j in ls:
            # print(row.find(j.lower()))
            try:
                dt[j] = row.find(j.lower()).text
            except:
                dt[j] = None
        # print(dt)
        res.append(dt)
    # display(pd.DataFrame(res))
    try:
        df = pd.DataFrame(res)
        df.columns = ['사용일자', '호선명', '역명', '승차송승객수', '하차총승객수']
        df.to_csv('./지하철/'+date+'.csv', encoding='euc-kr')
    except:
        print('데이터 없음')
print('end')
    
        

    
    
# items = soup.find_all('item')
# res = []
# ls = ['USE_DT', 'LINE_NUM', 'SUB_STA_NM', 'RIDE_PASGR_NUM', 'ALIGHT_PASGR_NUM']

# for item in items:
#     dt = {}
#     for j in ls:
#         try:
#             dt[j] = item.find(j).text
#         except:
#             dt[j] = None
#     res.append(dt)
# res
# pd.DataFrame(res)
```
