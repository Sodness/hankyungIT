```python
map=folium.Map(location=[37.5610336,126.9795475],zoom_start=11) # 서울시청을 중심으로 기본차트 작성

for i in range(len(df_top3)):
    자치구 = df_top3['자치구'][i]
    계 = df_top3['계'][i]
    
    folium.Marker(
        location=[df_top3['위도'][i], df_top3['경도'][i]],
        popup=f'<pre>{자치구}: {계:,}명</pre>',
        icon=folium.Icon(color='red',icon='fire')
    ).add_to(map)

map.choropleth(geo_data = geo_str,
               data = df,
               columns = ['자치구', '계'],
               fill_color = 'PuRd', #PuRd, YlGnBu
               key_on ='feature.id')
map
```
