```python
#User-gaent 
agent='Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36'
hdata={'User-Agent':agent}
html=requests.get(url, headers=hdata)
print(html.text)
```
