```python
res = [] #원본데이터 수집
exception_res = [] #이상데이터 수집
for i in df['분양가격']: #df['분양가격'] 정보 호출
    try: #error 상황 아닐 때 - int형으로 바꿀 수 있을 때
        if type(int(i)) == int:  #int형 바꿀 수 있으면
            res.append(i) #원본데이터만 수집
        else: #int형으로 바꿀 수 없을 때
            res.append(i) #원본데이터 수집
            exception_res.append(i) #이상데이터 수집
    except: #error 상황일 때
        res.append(i) #원본데이터 수집
        exception_res.append(i) #이상데이터 수집

set(exception_res) # 중복제거
```
