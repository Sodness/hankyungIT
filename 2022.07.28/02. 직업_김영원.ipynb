### Step 8. 직업별로 각 성별이 몇%를 차지하는지 구해주세요

administrator의 경우: 여성비율 45.56%, 남성비율 54.43%

```python
tmp = users1.groupby(['occupation', 'gender']).count()['age']
tmp
```

```python
tmp.unstack(1)
```

```python
tmp.unstack(1).fillna(0)
```

```python
tmp = tmp.unstack(1).fillna(0)
```

```python
tmp['F']
```

```python
여성비율 = tmp['F'] / (tmp['F']+tmp['M']) * 100
여성비율
```

```python
남성비율 = tmp['M'] / (tmp['F']+tmp['M']) * 100
남성비율
```

```python
users1.head()
```

```python
tmp = users1.groupby(['occupation', 'gender']).count()['age'].unstack(1)
tmp
```

```python
s = tmp.sum(axis=1)
s
```

```python
tmp['F_1'] = round((tmp['F'] / s * 100),2)
tmp['M_1'] = round((tmp['M'] / s * 100),2)
tmp
```
