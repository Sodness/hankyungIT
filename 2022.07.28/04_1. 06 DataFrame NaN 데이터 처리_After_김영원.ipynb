```python
import pandas as pd
import matplotlib.pyplot as plt
import missingno as msno 

train_data=pd.read_csv('./data/train.csv')

s_g=train_data.groupby('Survived')


s_g_mean=s_g.mean()['Age']

res=[]
for i, data in s_g:
    data['Age']=data['Age'].fillna(s_g_mean[i])
    res.append(data)



temp_res=pd.concat(res)

temp_res=temp_res.reset_index(drop=True)

display(temp_res.isnull().sum())
display(temp_res)
```
