## 상대 디렉토리 찾아가는 법  
. -> 현재 폴더 (현재 디렉토리)  
.. -> 상위 폴더 (상위 디렉토리)  
예시) ./images/ (뒤에 '/' 넣어줘야 폴더로 취급됨)  
<br>

## os 모듈 (디렉토리 관련)  
**// path에 있는 모든 파일목록을 리스트로 불러옴**  
os.listdir(주소)  
  
**// path에 'resize' 폴더를 생성**  
os.mkdir(주소 + 'resize')  
  
**// 해당 주소에 파일이 있는지 확인**  
os.path.isdir(주소/파일명)  
<br>

## 이미지만 다루는 모듈 pillow (다른 파일형식이 섞이면 오류 발생)

from PIL import Image  
  
**// 이미지 불러오기 Image.open("주소/파일명.확장자")**  
img = Image.open("./images/강아지1.jpg")

**// 이미지 속성 확인**  
img = Image.open("C:/Users/ilike/Pictures/IU_400x400.jpg")  
print(f'이미지 파일 이름 : {img.filename}')  
print(f'이미지 파일형식(format) : {img.format}')  
print(f'이미지 용량(size) : {img.size}')  
print(f'이미지 색상모드 : {img.mode}')  
print(f'이미지 width : {img.width}')  
print(f'이미지 height : {img.height}')  
  
**// 이미지 크기 변경 (resize)**  
img = Image.open("C:/Users/ilike/Pictures/IU_400x400.jpg")  
resize_img = img.resize((800, 800))  
resize_img.save("C:/Users/ilike/Pictures/IU_800x800.jpg") // 크기변경 이미지 저장할 주소 지정  
  
**// 이미지 자르기 (crop)**  
img = Image.open("C:/Users/ilike/Pictures/IU_400x400.jpg")  
cropped_img = img.crop((100, 100, 200, 200))  
cropped_img.show()  
  
**// 이미지 회전 (rotate)**  
img = Image.open("C:/Users/ilike/Pictures/IU_400x400.jpg")  
img.rotate(45).show()  
  
**// 이미지 대칭 (flip)**  
img = Image.open("C:/Users/ilike/Pictures/IU_400x400.jpg")  
flip_img = img.transpose(Image.FLIP_LEFT_RIGHT)  
flip_img.show()  
  
**// 이미지 필터링 (filter)**  
img = Image.open("C:/Users/ilike/Pictures/IU_400x400.jpg")
