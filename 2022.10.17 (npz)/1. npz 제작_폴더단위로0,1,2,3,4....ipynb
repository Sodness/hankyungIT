### 데이터셋제작
- class_names: 폴더명을 기록
- x값(이미지) : 폴더내의 모든 이미지를 읽어서 180,180 으로 사이즈변경후 제작
- Y값(레이블): clothes_dataste폴더의 폴더명을 순서대로 0,1,2,3으로 제작

```python
from glob import glob
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
folderList=glob('./clothes_dataset/*')
class_names=[]
for x in folderList:
    class_names.append(x.split('\\')[-1])
```

```python

class_names=os.listdir('./clothes_dataset/')
class_names
```

```python

x,y=[],[]
folderList=glob('./clothes_dataset/*')

for cnt,folder in enumerate(folderList):
    fileList=glob(folder + '/*.jpg')
    print('▶ Start:' + folder)
    for file in fileList:
        img=Image.open(file)
        img=img.resize((180,180))
        imgArr=np.array(img)
        x.append(imgArr)
        y.append(cnt)
        
```

```python
np.shape(x),np.shape(y)
```

```python
y[0]
```

```python
# 샘플출력
num=5
plt.imshow(x[num])
title=str(y[num])+ '-->'+ class_names[y[num]]
plt.title(title)
plt.show()
```

```python
y[445:459]   # black_dress 가 450개 이미지라 0번과 1번이 나타날수 있게끔
```

```python
np.savez('clothes.npz',  Xdata=x,  Ydata=y, class_names=class_names)
```
