```python
class Triangle:
    def __init__(self,b,h):
        self.b = b
        self.h = h
    
    def area(self):
        return self.b * self.h * (1/2)
```

# 다중상속시 부모클래스 접근

```python
class Car:
    def __init__(self):
        self.max_speed = 160
        self.speed = 0
        
    def speed_up(self, up_speed=20):
        if 160 <= self.speed:
            print('speed:', self.speed)
        else:
            self.speed = self.speed + up_speed
            print('speed:', self.speed)
    
    def speed_down(self, down_speed=20):
        if self.speed <= 0:
            print('speed:', self.speed)
        else:
            self.speed = self.speed - down_speed
            print('speed:', self.speed)

            
class SportCar(Car):
    def __init__(self):
        super(SportCar, self).__init__()
        self.max_speed = 200
        self.speed = 0
        
    def speed_up(self):
        super(SportCar,self).speed_up(up_speed=45)
    
    def speed_down(self):
        super(SportCar,self).speed_down(down_speed = 45)

            
class Truck(Car):
    def __init__(self):
        super(Truck, self).__init__()
        self.max_speed = 100
        self.speed = 0
        
    def speed_up(self):
        super(Truck,self).speed_up(up_speed=15)
    
    def speed_down(self):
        super(Truck,self).speed_down(down_speed = 15)
```
