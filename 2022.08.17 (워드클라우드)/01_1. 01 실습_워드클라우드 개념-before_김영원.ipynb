```python
txtList=['강아지 산책 강아지 목욕 강아지 미용 강아지 쇼핑',
 '친구와 저녁 먹음 가족과 점심 먹음 혼자 저녁 먹음 친구와 쇼핑']
```

```python
' '.join(txtList)
```

```python
txt = ' '.join(txtList)
```

```python
ls = ['강아지', '미용', '친구']
cnt_ls = []
for i in ls:
    cnt_ls.append(txt.count(i))

plt.figure(figsize=(15,5))
# 이렇게 for문 안돌려도 됨 그냥 리스트 넣어도 그려짐
for idx,v in enumerate(ls):
    plt.subplot(1,3,1)
    plt.bar(v, cnt_ls[idx])
    plt.title('워드빈도수 막대그래프')
    
    plt.subplot(1,3,2)
    plt.bar(v, cnt_ls[idx]/sum(cnt_ls))
    plt.title('워드빈도수 막대그래프(%)')
    
    plt.subplot(1,3,3)
    plt.pie(cnt_ls, labels=ls, autopct='%.1f%%')
    plt.title('워드빈도수 원차트')
```

```python
import requests
from bs4 import BeautifulSoup as bs
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family']='Malgun Gothic'
```

```python
url='https://movie.naver.com/movie/bi/mi/basic.naver?code=190991'
html = requests.get(url)
soup = bs(html.text, 'html.parser')

con_tx = soup.find_all('p', 'con_tx')
con_tx = [i.text for i in con_tx]
txt = " ".join(con_tx)
# print(txt)

word_ls = ['수학', '과학', '학문']
cnt_ls = [txt.count(i) for i in word_ls]
cnt_r = [i/sum(cnt_ls) for i in cnt_ls]

plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.bar(word_ls, cnt_ls)
plt.ylim(0,max(cnt_ls)+5)
plt.title('워드 빈도수 막대차트')

plt.subplot(1,3,2)
plt.bar(word_ls, cnt_r)
plt.title('워드 빈도수 %차트')

plt.subplot(1,3,3)
plt.pie(cnt_ls, labels=word_ls, autopct='%.1f%%')
plt.title('워드 빈도수 원차트')
plt.show()
```
