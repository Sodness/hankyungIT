```python
def token(okt_nouns, stopwords):
    return [i for i in okt_nouns if i not in stopwords]
```

```python
plt.figure(figsize=(20,20))

for idx,i in enumerate(sample['text']):
    wc = WordCloud("C:/windows/fonts/malgun.ttf",
                   background_color = "white",
                   max_words = 50,
                   stopwords = stopwords, 
                   width=300,
                   height=300)
    
    okt_nouns = okt.nouns(i)
    tmp = token(okt_nouns, stopwords)
    counter = Counter(tmp)
    
    wc = wc.generate_from_frequencies(counter) # {단어: 빈도수}를 기준으로 워드클라우드 생성, stopwords 안먹힘
    
    plt.subplot(1,5,idx+1)
    plt.imshow(wc)
    plt.axis('off')
```
