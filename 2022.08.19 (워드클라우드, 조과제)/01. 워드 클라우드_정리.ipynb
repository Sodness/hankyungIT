#### 워드 클라우드 (1) 
wc.generate(str자료)함수는 str형식으로 자료를 넣어줌

```python
#워드 클라우드
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np
# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
# 유니코드에서  음수 부호설정
plt.rc('axes', unicode_minus=False)


wc = WordCloud("C:/windows/fonts/malgun.ttf",
               background_color = "white",
               max_words = 2000)
               #stopwords = stopwords)
wc = wc.generate(str_words) # str자료만 가능, list자료는 안됨

#matplotlib로 시작화(1)
wc.to_image()
```

```python
#matplotlib로 시작화(2)
plt.figure(figsize = (12, 12))
plt.imshow(wc) #interpolation = "bilinear")
plt.axis("off")
plt.show()
```

#### 워드 클라우드 (2) 
`WordCloud.generate_from_frequencies`에 {단어:빈도수} 형식의 딕셔너리를 넣어주면 워드 클라우드가 생성됩니다

```python
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np

#단어빈도수 세기
sample=pd.read_csv('./data/전처리_샘플뉴스.csv')
okt=Okt()
words=okt.nouns(sample['text'][2]) 
words
```

```python
#단어 빈도수 세기
counter = Counter(words)
counter
```

```python
#워드 클라우드

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
# 유니코드에서  음수 부호설정
plt.rc('axes', unicode_minus=False)


wc = WordCloud("C:/windows/fonts/malgun.ttf",
               background_color = "white",
               max_words = 2000)
               #stopwords = stopwords)
wc = wc.generate_from_frequencies(counter) # str자료만 가능, list자료는 안됨

#matplotlib로 시작화(1)
wc.to_image()
```

### [미션] 기사별 워드 클라우드 표시 

```python
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np

#단어빈도수 세기
news=pd.read_csv('./data/전처리_샘플뉴스.csv')
okt=Okt()

#워드 클라우드 생성
wc = WordCloud("C:/windows/fonts/malgun.ttf",
               width=300,
               height=300,
               background_color = "white",
               max_words = 50)
               #stopwords = stopwords)
news_nouns=news['text'].apply(okt.nouns)
```

```python
news_nouns
```

```python
plt.figure(figsize=(30, 10))
for i, nouns in enumerate(news_nouns):
    counter = Counter(nouns)
    wc = wc.generate_from_frequencies(counter) 
    plt.subplot(1, 5, i+1)
    plt.imshow(wc)
    plt.axis('off')   

plt.show()
```
