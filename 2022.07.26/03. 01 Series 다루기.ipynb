###  * Series data 생성
> 1. pd.Series([1,2,3,4]) <br>
> 2. pd.Series([1,2,3,4], index=['a','b','c','d']

```python
# Series data 생성
a=pd.Series([1,2,3,4,])
print(a)
b=pd.Series([1,2,3,4], index=['a','b','c','d'])
print(b)
```

```python
#type()으로 확인
print(type(a))
print(type(b))
```

### * Series  속성
 > Series 속성에는 index 와values를 가집니다.

```python
print(b.index)
print(b.values)
print(a.index)
print(a.values)
```

### * 인덱싱 과 슬라이싱
> Series의 index번호와 index명으로 접근가능

```python
#인덱싱
print(b['b'])
print(b[1])

# 떨어진 2개이상의 인덱싱
print(b[['a','c']])
print(b[[1,3]])

#슬라이싱
print(b[0:2])
print(b[:-1])
print(b[-2:])
print(b['a':'c'])

# 값 바꾸기
b['a':'c']=100
print(b)
```

###  *  Series 함수
#### **Series size, shape, unique, count, value_counts 함수**
 - size : 개수 반환
 - shape : 튜플형태로 shape반환
 - unique: 유일한 값만 ndarray로 반환
 - count : NaN을 제외한 개수를 반환
 - mean: NaN을 제외한 평균 
 - value_counts: NaN을 제외하고 각 값들의 빈도를 반환 

```python
s=pd.Series([1,1,2,2,2,3,3,5,6,7,None])
s
```

```python
# len()
len(s)
```

```python
#s.size
s.size
```

```python
# s.shape
s.shape
```

```python
#s.unique()
s.unique()
```

```python
#s.count()
s.count
```

```python
# s.value_counts()
s.value_counts()
```
