```python
# 데이터프레임에서 원하는 데이터 추출 4가지

# []: '열이름' 기반 필터링
# >=: boolean 필터링
# .loc[,]: '열이름' 기반 인덱싱
# .iloc[,]: index 기반 인덱싱

df = pd.read_csv('./data/cctv_in_seoul.csv')
df.head()
```

```python
# []: '열이름' 기반 필터링
df[['기관명']]
```

```python
df[['기관명', '소계']]
```

```python
df[:3]
```

```python
# >=: boolean 필터링
df['2014년'] >= 100
```

```python
df[df['2014년'] >= 100]
```

```python
# 조건이 여러개 일 때 : &(and),  | (or)
cond1=df['2015년']>=200
cond2=df['2016년']>=200
df[cond1 & cond2]
```

```python
# .loc[,]: '열이름' 기반 인덱싱
df.loc[1:4]
```

```python
df.loc[1:4, ['기관명', '소계']]
```

```python
# .iloc[,]: index 기반 인덱싱
df.iloc[[1]]
```

```python
df.iloc[:3, -3:]
```

```python
# set_index('열이름'): index를 '기관명'으로 변경
# reset_index(): 뒤섞인 인덱스를 정렬
df.set_index('기관명')
```

```python
# df['열이름'].str.contains('찾는 문자')

df['기관명'].str.contains('강')
```

```python
df[df['기관명'].str.contains('강')]
```
