```python
# 언더바 2개여서 내 코드로는 안되네
# 아니네 내꺼도 되네
src_ls = glob('./BTS_MV_YOLO/labels/*')
src_ls = [i.split('\\')[-1] for i in src_ls]

dst_ls = glob('./BTS_MV_YOLO/labels/*')
dst_ls = [i.split('_')[-1] for i in dst_ls]
dst_ls = [int(i.split('.')[0]) for i in dst_ls]

dst_ls.sort()
max_len = len(str(dst_ls[-1]))
tmp = ''
for i in range(max_len-1):
    tmp = tmp + '0'

dst_ls = [str(i) for i in dst_ls]
dst_ls = [tmp+i for i in dst_ls]
dst_ls = [i[-max_len:] for i in dst_ls]

path = './BTS_MV_YOLO/labels/'
for i,item in enumerate(dst_ls):
    src = os.path.join(path, src_ls[i])
    dst = 'BTS_MV_' + item + '.txt'
    dst = os.path.join(path, dst)
    os.rename(src, dst)
```

```python
fileList = glob("./BTS_MV_YOLO/labels/*.txt")

res = []
for file in fileList:
    f = open(file, 'r')
    lines = f.readlines()
    for line in lines:
        line = line.strip()  # 줄 끝의 줄 바꿈 문자를 제거한다.
        # print(line)
        res.append(line.split(' ')[0])
    f.close()
res
```

```python
from collections import Counter

counter = Counter(res)

# print(res)
# print(counter)
print(dict(counter))
```

```python
from glob import glob
import numpy as np
import matplotlib.pyplot as plt

fileList = glob('./BTS_MV_YOLO/labels/*.txt')
videoinfo = []

for x in fileList:
    f = open(x,'r')
    lines = f.readlines()

    for line in lines:
        videoinfo.append([])
        videoinfo[-1].append(line.split(' ')[0])
        videoinfo[-1].append(x[-7:-4])

videoinfo = np.array(videoinfo)
videoCount = np.unique(videoinfo[:,0], return_counts=True)

plt.barh(videoCount[0], videoCount[1])
```
