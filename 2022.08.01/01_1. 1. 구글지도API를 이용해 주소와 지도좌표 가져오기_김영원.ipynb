```python
df = pd.read_excel('./crime_2017_1.xls')
df
```

```python
rename = '서울시 ' + df['자치구']
rename
```

```python
res = []

for v in rename:
    tmp = gmaps.geocode(v, language='ko')
    dic = {
        '주소': tmp[0]['formatted_address'],
        '위도': tmp[0]['geometry']['location']['lat'],
        '경도': tmp[0]['geometry']['location']['lng']
    }
    res.append(dic)
res
```

```python
df1 = pd.DataFrame(res)
df1
```

```python
df = pd.concat([df, df1], axis=1)
df
```
