```python
import cv2


#  
cap = cv2.VideoCapture(0)
# Lists to store the bounding box coordinates

top_left_corner = []

bottom_right_corner = []

# function which will be called on mouse input


def drawRectangle(action, x, y, flags, parms):


    # Referencing global variables
    global top_left_corner, bottom_right_corner,ret, image
    # Mark the top left corner when left mouse button is pressed
    if action == cv2.EVENT_LBUTTONDOWN:
        top_left_corner = [(x, y)]
    # When left mouse button is released, mark bottom right corner
    elif action == cv2.EVENT_LBUTTONUP:
        bottom_right_corner = [(x, y)]
        # Draw the rectangle
        cv2.rectangle(image, top_left_corner[0], bottom_right_corner[0], (0, 255, 0),5)
        while True:
            ret, image = cap.read() #     ret-> True, frame->  

            if not ret:
                break

            cv2.imshow('Window', image)
            cv2.setMouseCallback("Window", drawRectangle)
            if cv2.waitKey(1) == 27:  #  (,10ms)    ,  ESC(27)  break
                break
        cap.release() #   
        cv2.destroyAllWindows()
        


#   
while True:
    ret, image = cap.read() #     ret-> True, frame->  

    if not ret:
        break
    
    cv2.imshow('Window', image)
    cv2.setMouseCallback("Window", drawRectangle)
    if cv2.waitKey(1) == 27:  #  (,10ms)    ,  ESC(27)  break
        break
cap.release() #   
cv2.destroyAllWindows()
```

```python
import cv2

cap = cv2.VideoCapture(0)

fourcc = cv2.VideoWriter_fourcc(*'DIVX') # 

out = cv2.VideoWriter('output.avi', fourcc, 25.0, (640,480)) #   

while True:
    ret, frame = cap.read() #     ret-> True, frame->  

    if not ret:
        break

    out.write(frame)
    cv2.imshow('frame', frame)

    if cv2.waitKey(10) == ord('q'):
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
import cv2

cap = cv2.VideoCapture(0)

fourcc = cv2.VideoWriter_fourcc(*'DIVX') # 

out = cv2.VideoWriter('output.avi', fourcc, 25.0, (640,480)) #   

while True:
    ret, frame = cap.read() #     ret-> True, frame->  

    if not ret:
        break

    inversed = 1-frame  # 
    out.write(inversed)
    cv2.imshow('frame', frame)
    cv2.imshow('inversed',inversed)

    if cv2.waitKey(10) == ord('q'):
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```
