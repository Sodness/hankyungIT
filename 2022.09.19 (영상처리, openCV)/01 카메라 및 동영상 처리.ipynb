### 1. 카메라



```python
import cv2


# 카메라 열기
cap = cv2.VideoCapture(0)
# Lists to store the bounding box coordinates

top_left_corner = []

bottom_right_corner = []

# function which will be called on mouse input


def drawRectangle(action, x, y, flags, parms):


    # Referencing global variables
    global top_left_corner, bottom_right_corner,ret, image
    # Mark the top left corner when left mouse button is pressed
    if action == cv2.EVENT_LBUTTONDOWN:
        top_left_corner = [(x, y)]
    # When left mouse button is released, mark bottom right corner
    elif action == cv2.EVENT_LBUTTONUP:
        bottom_right_corner = [(x, y)]
        # Draw the rectangle
        cv2.rectangle(image, top_left_corner[0], bottom_right_corner[0], (0, 255, 0),5)
        while True:
            ret, image = cap.read() #카메라로부터 정상적으로 프레임을 받아 오면 ret-> True, frame-> 해당 프레임저장

            if not ret:
                break

            cv2.imshow('Window', image)
            cv2.setMouseCallback("Window", drawRectangle)
            if cv2.waitKey(1) == 27:  # 일정시간 (예,10ms) 후 다음 프레임 읽음, 만약 ESC(27)키를 누르면 break
                break
        cap.release() # 사용한 자원 해제
        cv2.destroyAllWindows()
        


# 카메라 프레임 처리
while True:
    ret, image = cap.read() #카메라로부터 정상적으로 프레임을 받아 오면 ret-> True, frame-> 해당 프레임저장

    if not ret:
        break
    
    cv2.imshow('Window', image)
    cv2.setMouseCallback("Window", drawRectangle)
    if cv2.waitKey(1) == 27:  # 일정시간 (예,10ms) 후 다음 프레임 읽음, 만약 ESC(27)키를 누르면 break
        break
cap.release() # 사용한 자원 해제
cv2.destroyAllWindows()
```

### 2. 동영상 파일 

```python
import cv2


# 카메라 열기
cap = cv2.VideoCapture('./video1.mp4') #장치ID가 아닌 동영상 파일명만 바꾸면 됨



# 카메라 프레임 처리
while True:
    ret, frame = cap.read() #카메라로부터 정상적으로 프레임을 받아 오면 ret-> True, frame-> 해당 프레임저장

    if not ret:
        break
    
    cv2.imshow('frame', frame)
    if cv2.waitKey(20) == 27:  # 일정시간 (예,10ms) 후 다음 프레임 읽음, 만약 ESC(27)키를 누르면 break
        break
cap.release() # 사용한 자원 해제
cv2.destroyAllWindows()
```

### 3. 동영상 저장






```python
import cv2

cap = cv2.VideoCapture(0)

fourcc = cv2.VideoWriter_fourcc(*'DIVX') #코덱 설정

out = cv2.VideoWriter('output.avi', fourcc, 25.0, (640,480)) #영상을 저장할 객체 생성

while True:
    ret, frame = cap.read() #카메라로부터 정상적으로 프레임을 받아 오면 ret-> True, frame-> 해당 프레임저장

    if not ret:
        break

    out.write(frame)
    cv2.imshow('frame', frame)

    if cv2.waitKey(10) == ord('q'):
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
import cv2

cap = cv2.VideoCapture(0)

fourcc = cv2.VideoWriter_fourcc(*'DIVX') #코덱 설정

out = cv2.VideoWriter('output.avi', fourcc, 25.0, (640,480)) #영상을 저장할 객체 생성

while True:
    ret, frame = cap.read() #카메라로부터 정상적으로 프레임을 받아 오면 ret-> True, frame-> 해당 프레임저장

    if not ret:
        break

    inversed = 1-frame  # 반전
    out.write(inversed)
    cv2.imshow('frame', frame)
    cv2.imshow('inversed',inversed)

    if cv2.waitKey(10) == ord('q'):
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```
